```python
import re

with open('New_00_modified_final.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# 使用正则表达式匹配每一行中的“Графіті № (数字) (табл. (罗马数字), (数字))”格式
pattern = re.compile(r'Графіті № (\d+) \(табл\. ([IVXLCDM]+), (\d+)\)')

# 将匹配到的信息存入不同的变量
grafiti_numbers = []
table_roman_numerals = []
table_numbers = []

for line in lines:
    match = pattern.search(line)
    if match:
        grafiti_numbers.append(match.group(1))
        table_roman_numerals.append(match.group(2))
        table_numbers.append(match.group(3))

# 打印结果或进行其他操作
for i in range(len(grafiti_numbers)):
    print(f"Графіті № {grafiti_numbers[i]}, табл. {table_roman_numerals[i]}, {table_numbers[i]}")
```

```python

with open('New_00_modified_final.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

pattern = re.compile(r'Графіті №')

# 将匹配到的信息存入不同的变量
matches = []

for line in lines:
    match = pattern.search(line)
    if match:
        grafiti_info = match.group(0)
        matches.append(grafiti_info)

# 打印匹配到的信息或进行其他操作
for match in matches:
    print(match)
```

```python
pat = '[IVXLCDM]+'
if re.search(pat, 'VI'):
    print('Valid Roman')
```

```python
import re
pattern = '^M{0,3}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$'
if re.search(pattern, 'I'):
    print('Valid Roman')
else:
    print('Not valid Roman')
```

```python
import re

with open('New_00_modified_final.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# 使用正则表达式以"Графіті № "为分隔符，提取每个部分
pattern = re.compile(r'Графіті № ')
sections = pattern.split(text)

# 打印每个提取的部分或进行其他操作
for section in sections:
    print(section.strip())  # 去除两端的空白，如果需要
```
