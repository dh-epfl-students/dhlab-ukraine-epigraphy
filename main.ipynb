```python
import re
import json

# 读取文本文件
with open('New_00_Part_1_clean.txt', 'r', encoding='utf-8') as file:
    content = file.readlines()

# 初始化存储数据的字典
data = {}
current_part = None
current_section = None
current_subsection = None

# 用于匹配需要插入换行符的文本的正则表达式
part_pattern = re.compile(r'^ГРАФІТІ')
section_pattern = re.compile(r'^I{1,3}\.\d+\.\sГрафіті')
subsection_pattern = re.compile(r'^Графіті №')

# 遍历每一行文本内容
for line in content:
    line = line.strip()
    
    # 忽略空行
    if not line:
        continue
    
    # 检查是否匹配到新的部分
    if re.match(part_pattern, line):
        current_part = line
        data[current_part] = {"description": ""}
        current_section = None
        current_subsection = None
    # 检查是否匹配到新的章节
    elif re.match(section_pattern, line):
        current_section = line
        data[current_part][current_section] = {"about": "", "subsections": []}
        current_subsection = None
    # 检查是否匹配到新的子章节
    elif re.match(subsection_pattern, line):
        current_subsection = line
        data[current_part][current_section]["subsections"].append({"name": current_subsection, "content": {}})
    # 检查关键词并将内容存入相应标签中
    elif current_part:
        if current_section and current_subsection:
            for keyword in ["Розташування", "Опис", "Датування", "Категорія"]:
                if line.startswith(keyword):
                    key = keyword
                    value = line[len(keyword):].strip()
                    data[current_part][current_section]["subsections"][-1]["content"][key] = value
        elif current_section:
            data[current_part][current_section]["about"] += line + "\n"
        else:
            data[current_part]["description"] += line + "\n"

# 将数据写入JSON文件
with open('New_00_Part_1_clean.json', 'w', encoding='utf-8') as json_file:
    json.dump(data, json_file, ensure_ascii=False, indent=4)
```

```python
import re

with open('New_00_modified_final.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# 使用正则表达式以"Графіті № "为分隔符，提取每个部分
pattern = re.compile(r'(?=Графіті № )')
sections = pattern.split(text)

cleaned_sections = [section.replace('\n', ' ').strip() for section in sections]
cleaned_sections = [cleaned_section for cleaned_section in cleaned_sections if cleaned_section.startswith('Графіті №')]
for cleaned_section in cleaned_sections:
    
    print(cleaned_section.strip())  # 去除两端的空白，如果需要
```

```python
cleaned_sections
```

```python
modified_sections = []
for cleaned_section in cleaned_sections:
    # 在"Графіті №"后增加句点"."
    modified_section = re.sub(r'Графіті №', r'Графіті №.', cleaned_section)
    modified_sections.append(modified_section)

# 打印处理后的结果
for modified_section in modified_sections:
    print(modified_section)
```

```python
keywords = ["Графіті №","табл", "Публікації", "Публікація", "Розташування", "Опис", "Датування", "Категорія"]
```

```python
import pandas as pd
df = pd.DataFrame(columns=keywords)
for modified_section in modified_sections:
    row = {}
    for keyword in keywords:
        # 使用正则表达式匹配关键词后的内容
        match = re.search(f'{keyword}\.(.*?)(?={ "|".join(keywords) }|$)', modified_section)
        if match:
            row[keyword] = match.group(1).strip()
        else:
            row[keyword] = None

    # 将提取的信息添加到DataFrame
    df = df.append(row, ignore_index=True)

```

```python
df
```

```python
def remove_last_char(s):
    a = str(s)
    a = a[:-1]
    return a

# 对 "Графіті №" 列应用函数
df['Графіті №'] = df['Графіті №'].apply(remove_last_char)
df['табл'] = df['табл'].apply(remove_last_char)
```

```python
def IOB_transform(sentence):
    doc = nlp(str(sentence))
    iob_tags = [(token.text, token.ent_iob_) for token in doc if token.text not in ['|', ',', '.', ':', ',', '[', ']', '(', ')']]
    iob_tags_no_stop_words = [element for element in iob_tags if iob_tags[0] not in stopwords_list]
    return iob_tags_no_stop_words
```

```python
def extract_terms(text):
    doc = nlp(str(text))
    terms = []
    for token in doc:
        if token.pos_ in ["NOUN", "PROPN"]:
            terms.append(token.text)
    return terms
```

```python
terms_dict = dict(zip(terms['keyword'], terms['label-IOB']))

print(terms_dict)
```

```python
import spacy

nlp = spacy.load("uk_core_news_lg")
for key in data_dict.keys():
    part = data_dict[key]
    description = part['description']
    doc = nlp(description)
```

```python
from spacy.tokens import Span
for keyword, label in terms_dict.items():
    nlp.vocab.strings.add(label)

@spacy.Language.component("custom_ner")
def custom_ner_component(doc):
    spans = []
    for i in range(len(doc)):
        if doc[i].text.lower() in terms_dict:
            span = Span(doc, i, i+1, label=terms_dict[doc[i].text.lower()])
            spans.append(span)
    doc.ents = spans
    return doc

```

```python
for token in doc:
    print(token.text)
```

```python
file_path = "New_00_Part_1.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()

# 分词
doc = nlp.tokenizer(text)

# 打印分词结果
word_dict = []
for token in doc:
    if not token.is_punct:  # 检查是否是标点符号
        if token.text in terms_dict.keys():
            word_dict.append([token.text, terms_dict[token.text]])
        else:
            word_dict.append([token.text, 'O'])
```
